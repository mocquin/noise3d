```python
import noise3d
import noise3d.stats
```

```python
T = V = H = 100
seq = noise3d.genseq.genseq_tvh(T, V, H, 0, 1)
```

# Basic statistics

```python
noise3d.stats.basic_stats(seq)
```

# Confidence intervals

```python
# raw normal CI
noise3d.stats.compute_CI(seq, print_CI=True)
```

```python
# Improve CI if mean and std are known
noise3d.stats.compute_CI(seq, mu_X=0, sigma_X=1, print_CI=True)
```

```python
# change the alpha parameter : you get smaller CI but with less chance the actual value is in this CI
noise3d.stats.compute_CI(seq, print_CI=True, alpha=0.5)
```
