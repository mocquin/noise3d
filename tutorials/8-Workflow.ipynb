```python
import noise3d
import numpy as np
```

# Typical workflow example

## Generate data

Suppose you collect a sequence of noise, that we will simulate here.

```python
T = V = H = 100
# pick 7 noise std, uniformly between 0 and 2
sigmas = np.random.uniform(0, 2, 7)
print("Random data generated with standard deviations: ")
for s, n in zip(sigmas,noise3d.NAMES):
    print(f"{n:3} : {s:.3f}")
# generate random data
data = noise3d.genseq_3dnoise_seq(T, V, H, sigmas)
```

## Quick noise estimation

The first thing we want to do is to have a quick estimation of the different noises.

```python
noise3d.disp.display_8seq(data, extract=True, samex=True);
```
