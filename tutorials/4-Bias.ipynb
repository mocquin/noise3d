# Bias

The "classic" methods of estimation of the variances are biased. It can be demonstrated analyticaly, and the error can be computed.
The bias module aims to demonstrate this effect, by comparing : 
 - the analytical errors
 - the actual error measured on simulated data
 
All the work is wrapped up in a class.

```python
demo = bias.BiasedEstimatorTVH()
```

```python
demo.setup()
demo.compute_results()
```
